```python
from ludwig import LudwigModel
import pandas as pd
```

```python
ludwig_columns = ['No', 'type', 'area_name', 'area_code', 'states', 'city', 'district', 'closest_station',
       'closest_station_distance', 'value', 'price_per_of_area', 'floor_plan', 'area', 'transaction_price', 'shape',
       'frontage', 'total_floor_area', 'year_of_construction', 'construction', 'use', 'future_use_purpose', 'orientation_front_road', 'type_front_road',
       'road_width', 'city_planning', 'building_coverage', 'floor_area_ratio', 'time_of_transaction', 'renovation', 'transaction_circumstances']
```

```python
train_df = pd.read_csv('32_Shimane Prefecture_20171_20182.csv', names=ludwig_columns, header=0, index_col="No")
test_df = pd.read_csv('32_Shimane Prefecture_20183_20183.csv', names=ludwig_columns, header=0, index_col="No")
train_df.columns
```

```python
train_df.to_csv('shimane_train_en.csv')
test_df.to_csv('shimane_test_en.csv')
```

```python
model_definition ={
    "input_features": [
        {"name": "type", "type": "category"},
        {"name": "area_name", "type": "category"},
        {"name": "area_code", "type": "category"},
        {"name": "closest_station", "type": "category"},
        {"name": "future_use_purpose", "type": "category"},
        {"name": "building_coverage", "type": "numerical"},
        {"name": "road_width", "type": "numerical"},
        {"name": "renovation", "type": "category"},
    ],
    "output_features": [{"name": "value", "type": "numerical"}],
}
```

```python
model = LudwigModel(model_definition)
```

```python
train_stats = model.train(train_df)
```

```python
predictions = model.predict(test_df)
```
